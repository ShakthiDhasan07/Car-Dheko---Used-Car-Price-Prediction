# Outliers

```python
sns.distplot(x=df_car["price"], kde=True)  
plt.title("Distribution of Price")
plt.show()
```

```python
df_car["price"]
```

```python
Q1=df_car["price"].quantile(0.25)
Q3=df_car["price"].quantile(0.75)

IQR=Q3-Q1
```

```python
df_final_car=df_car[(df_car["price"]>=(Q1-1.5*IQR)) & (df_car["price"]<=(Q3+1.5*IQR))]
```

```python
lower_outlier = len(df_car[df_car['price'] < (Q1 - 1.5 * IQR)])

# Count of upper outliers
upper_outlier = len(df_car[df_car['price'] > (Q3 + 1.5 * IQR)])
```

```python
print(lower_outlier)
print(upper_outlier)
```

```python
df_final_car.reset_index(inplace=True,drop=True)
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Create a list to hold the name of the categorical columns
catg_col = df_final_car.select_dtypes(include="object").columns.to_list()

# Initialize a dictionary to hold the encoders for each column
label_encoders = {}

# Apply Label Encoding to each categorical column
for col in catg_col:
    encoder = LabelEncoder()
    # Apply fit_transform on each column and store it back in the DataFrame
    df_final_car[col] = encoder.fit_transform(df_final_car[col])
    
    # Store the encoder in the dictionary
    label_encoders[col] = encoder

# Check the result - print the first few rows to see the transformed data
print(df_final_car.head())

# Now check if the encoders are populated correctly

```
